```python
import requests
from bs4 import BeautifulSoup
import time

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36"
}

start_url = "https://www.ajnet.me"
visited = set()
to_visit = [start_url]
i = 0
while to_visit:
    url = to_visit.pop(0) 
    if url in visited:
        continue
    try:
       response = requests.get(url, headers=headers)
       soup = BeautifulSoup(response.content, "html.parser")
       visited.add(url)

       print(f"\n{i}:Visiting: {url}")
    

    
       for link_tag in soup.find_all("a", href=True):
            href = link_tag['href']
       
            if href.startswith("/"):
                full_link = "https://www.ajnet.me" + href
            elif href.startswith("https://www.ajnet.me/"):
                full_link = href
            else:
                continue

            if full_link not in visited:
                to_visit.append(full_link)
                
       i+= 1
       time.sleep(1)

    except Exception as e:
       print(f"Error visiting {url}: {e}")
```

```python
import numpy as np
urls = np.array(list(visited))
urls.size
```

```python
import pandas as pd
DF = pd.DataFrame(urls) 
DF
DF.to_csv(f"C:\\Users\\karem\\OneDrive\\Desktop\\linkes_ajnet.csv")
```
