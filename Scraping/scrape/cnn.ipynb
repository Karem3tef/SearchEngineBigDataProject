```python
import requests
from bs4 import BeautifulSoup
import time

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36"
}

start_url = "https://edition.cnn.com" 
visited = set()
to_visit = [start_url]
i = 0  

while to_visit :
    url = to_visit.pop(0)

    if url in visited:
        continue

    try:
        response = requests.get(url, headers=headers, timeout=10)
        soup = BeautifulSoup(response.content, "html.parser")
        visited.add(url)

        title = soup.title.string if soup.title else "No Title"
        print(f"\n[{i}] Visiting: {url}")
        print(f"Page Title: {title}")

        for link_tag in soup.find_all("a", href=True):
            href = link_tag['href']

            if href.startswith("/"):
                full_link = start_url + href
            elif href.startswith("https://edition.cnn.com"):
                full_link = href
            else:
                continue

            if full_link not in visited and full_link not in to_visit:
                to_visit.append(full_link)

        i += 1
        time.sleep(1)

    except Exception as e:
        print(f"Error visiting {url}: {e}")
```

```python
import pandas as pd   
import numpy as np
urls = np.array(list(visited))
print(urls.size)
print(urls) 
```

```python
DF = pd.DataFrame(urls) 
DF
DF.to_csv(f"C:\\Users\\karem\\OneDrive\\Desktop\\linkes.csv")
```
