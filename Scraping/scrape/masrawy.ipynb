```python
import requests
from bs4 import BeautifulSoup
import time



start_url = "https://www.masrawy.com"
visited = set()
to_visit = [start_url]
i = 0 

while to_visit :
    url = to_visit.pop(0)

    if url in visited:
        continue

    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")

        visited.add(url)

        title = soup.title.string.strip() if soup.title else "No Title"
        print(f"\n[{i}] Visiting: {url}")
        print(f"Page Title: {title}")

        for link_tag in soup.find_all("a", href=True):
            href = link_tag['href']

            if href.startswith("/"):
                full_link = start_url + href
            elif href.startswith("https://www.masrawy.com"):
                full_link = href
            else:
                continue

            if full_link not in visited and full_link not in to_visit:
                to_visit.append(full_link)

        i += 1
        time.sleep(1)

    except Exception as e:
        print(f"Error visiting {url}: {e}")
```

```python
import numpy as np
urls = np.array(list(visited))
urls.size
```
